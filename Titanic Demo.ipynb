```python
# suppress warning messages
import warnings; warnings.simplefilter('ignore')
```

```python
# import libraries
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#enable inline plot
%matplotlib inline
```

```python
# use pandas library to read .csv file of passengers on titanic and store in a dataframe object
titanic = pd.read_csv('data/train.csv')
```

```python
# check first few lines of dataframe
titanic.head()
```

```python
# get basic information about the dataframe
titanic.info()
```

```python
# get counts by sex or gender
titanic['Sex'].value_counts()
```

```python
# get counts by survival
titanic['Survived'].value_counts()
```

```python
# plot survival counts
sns.factorplot('Survived',data=titanic,kind='count')
```

```python
# plot survivability by cabin class
sns.factorplot('Pclass','Survived',data=titanic)
```

```python
# plot survivability by sex
sns.factorplot('Sex','Survived',data=titanic)
```

```python
# plot survivability by sex and broken down by cabinclass
sns.factorplot('Sex','Survived',hue='Pclass',data=titanic)
```

```python
# plot histogram of age
titanic['Age'].hist(bins=50)
```

```python
# overlay histogram with kde
sns.distplot(titanic['Age'].dropna())
```

```python
#use FacetGrid to plot multiple kdeplots of Age and Sex on one plot 
fig = sns.FacetGrid(titanic,hue='Sex',aspect=4)
fig.map(sns.kdeplot,'Age',shade=True)
oldest = titanic['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
#use FacetGrid to plot multiple kdeplots of Age and Pclass on one plot
fig = sns.FacetGrid(titanic,hue='Pclass',aspect=4)
fig.map(sns.kdeplot,'Age',shade='True')
oldest = titanic['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```
